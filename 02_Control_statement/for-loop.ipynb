```python
jackpot = 12

guess = int(input("Chal guess kar"))
counter = 1

while guess != jackpot:
    if guess < jackpot:
        print("Guess higher")
    else:
        print("Guess lower")
        
    guess = int(input("Chal guess kar"))
    counter+=1
    
    
print("Sahi Jawab")
print("You took",counter,"attempts")
    
```

nested-loops

```python
rows = int(input("Enter the number of rows "))

for i in range(1,rows + 1):
    for j in range(0,i):
        print("*",end=" ")
    print("")
```
